```python
import re

# hangi text kolonunu kullanıyorsan onu yaz (abstract, text, title vs.)
text_cols = ["url", "text"]   # gerekirse değiştir

# url + text'i birleştirip lower alıyoruz
s = (df_news[text_cols].fillna("").agg(" ".join, axis=1).str.lower())

mask = (
    s.str.contains("artificial_intelligence") |
    s.str.contains("machine_learning") |
    s.str.contains("deep_learning") |
    s.str.contains("neural_network") |
    s.str.contains("autonomous") |
    s.str.contains("robotics") |
    s.str.contains("big_data") |
    s.str.contains("internet_of_things") |
    s.str.contains("intelligent_systems") |
    s.str.contains("data-driven") |
    s.str.contains("large-scale-dataset") |
    s.str.contains("smart_objects") |
    s.str.contains(r"\bai\b", regex=True) |
    s.str.contains(r"\bai[- ]", regex=True) |
    s.str.contains(r"\biot\b", regex=True)
)

# sadece eşleşen satırlar kalsın
df_news = df_news[mask].reset_index(drop=True)
df_news
```

```python
import re

ai_terms = [
    "artificial_intelligence",
    "machine_learning",
    "deep_learning",
    "neural_network",
    "autonomous",
    "robotics",
    "big_data",
    "internet_of_things",
    "intelligent_systems",
    "data-driven",
    "large-scale-dataset",
    "smart_objects",
    "ai",      # regex ile ayrı ele alacağız
    "iot"
]

def extract_tech_field(theme_str):
    if not isinstance(theme_str, str):
        return None
    
    t = theme_str.lower()
    
    # regex bazlı özel aramalar
    if re.search(r"\bai\b", t):
        return "ai"
    if re.search(r"\biot\b", t):
        return "iot"
    
    # normal liste aramaları
    for term in ai_terms:
        if term in t:
            return term
    
    return None

def title_from_url(url):
    if not isinstance(url, str):
        return None

    url = url.rstrip('/')

    slug = url.split('/')[-1]

    # slug tamamen sayı ise (ID ise) başlık çıkmaz
    if slug.isdigit():
        return None

    # başı sayı olan slug (70301-ibm-to-spend...) → sayıyı düşür
    parts = slug.split('-')
    if parts[0].isdigit() and len(parts) > 1:
        slug = "-".join(parts[1:])

    # alfanumerik ve tire dışındaki karakterleri temizle
    slug = re.sub(r"[^a-zA-Z0-9-]", "", slug)

    title = slug.replace('-', ' ').strip().title()

    return title if title else None
```

```python
df = df_news.copy()

df_news_corpus = pd.DataFrame()
df_news_corpus["title"] = df_news["url"].apply(title_from_url)
df_news_corpus["abstract"]        = df_news["text"].where(df_news["text"].notna(), None)
df_news_corpus["detailed_text"]= df_news["text"].where(df_news["text"].notna(), None) + " " + df_news_corpus["title"]
df_news_corpus["year"]        = df_news["year"]
df_news_corpus["month"]       = df_news["month"]
df_news_corpus["source_type"] = "news"
df_news_corpus["tech_field"] = df_news["themes"].apply(extract_tech_field)
df_news_corpus["trl"] = "NaN"


df_news_corpus
```

```python
out_path = "../data/news_ai_corpus_2015_2025.csv"
df_news_corpus.to_csv(out_path, index=False)
```
