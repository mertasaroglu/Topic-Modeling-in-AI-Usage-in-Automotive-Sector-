```python
import os
import zipfile
import shutil

DATASET = "mertasaroglu/nasa-trl-labeling-ia-in-automotive-sector"

# 1) Hedef klasörleri ayarla
ROOT = os.path.abspath("../..")  # notebook içindeki konumuna göre ayarla
TARGET_DIR = os.path.join(ROOT, "01_data", "predictive_model")

os.makedirs(TARGET_DIR, exist_ok=True)

# 2) ZIP dosyasının adı Kaggle formatına göre
ZIP_NAME = "nasa-trl-labeling-ia-in-automotive-sector.zip"

print(">> Downloading Kaggle dataset...")
!kaggle datasets download -d {DATASET} -p "{TARGET_DIR}"

# 3) ZIP tam yolu
zip_path = os.path.join(TARGET_DIR, ZIP_NAME)

# 4) ZIP dosyasını aç
print(">> Extracting ZIP...")
with zipfile.ZipFile(zip_path, "r") as z:
    z.extractall(TARGET_DIR)

# 5) ZIP’i sil
os.remove(zip_path)

print(">> Done. Extracted files:")
print(os.listdir(TARGET_DIR))
```
