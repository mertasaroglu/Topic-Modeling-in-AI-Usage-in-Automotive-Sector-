```python
import pandas as pd
```

```python
df_paper=pd.read_csv("../functions/paper_ai_master_2015_2025.csv")

df_paper.head()
```

```python
df_paper.info()
```

```python
df = df_paper.copy()

df_paper_corpus = pd.DataFrame()
df_paper_corpus["title"] = df_paper["title"].where(df_paper["title"].notna(), None)
df_paper_corpus["abstract"]  = df_paper["text"].where(df_paper["text"].notna(), None)
df_paper_corpus["detailed_text"]= df_paper["text"].where(df_paper["text"].notna(), None) + " " + df_paper_corpus["title"]
df_paper_corpus["year"]        = df_paper["year"]
df_paper_corpus["month"]       = df_paper["month"]
df_paper_corpus["source_type"] = df_paper["source_type"]
df_paper_corpus["tech_field"] = df_paper["tech_field"].where(df_paper["tech_field"].notna(), None)
df_paper_corpus["trl"] = "NaN"


df_paper_corpus
```

```python
df_paper_corpus.info()
```

```python
out_path = "../data/paper_ai_corpus_2015_2025.csv"
df_paper_corpus.to_csv(out_path, index=False)
```
